```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Read the image
img = cv2.imread('download.jfif')

average_kernel = np.ones((5,5), np.float32) / 25
gaussian_kernel = cv2.getGaussianKernel(5, -1) @ cv2.getGaussianKernel(5, -1).T

average_filtered = cv2.filter2D(img, -1, average_kernel)
gaussian_filtered = cv2.filter2D(img, -1, gaussian_kernel)

plt.figure(figsize=(10,10))
plt.subplot(131), plt.imshow(img), plt.title('Original')
plt.subplot(132), plt.imshow(average_filtered), plt.title('Average Filter')
plt.subplot(133), plt.imshow(gaussian_filtered), plt.title('Gaussian Filter')
plt.show()
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Read the image with salt and pepper noise
img = cv2.imread('download.jfif')

# Apply a median filter (rank=13)
median_filtered = cv2.medianBlur(img, 5)

# Display the original and filtered images
plt.figure(figsize=(10,10))
plt.subplot(121), plt.imshow(img), plt.title('Original Image with Noise')
plt.subplot(122), plt.imshow(median_filtered), plt.title('Median Filter (Rank 13)')
plt.show()
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Read the image
img = cv2.imread('download.jfif', 0)

# Function to manually implement Laplacian filter
def mylaplacian(img):
    vertical_mask = np.array([[0, 1, 0], [0, -2, 0], [0, 1, 0]])
    horizontal_mask = np.array([[0, 0, 0], [1, -2, 1], [0, 0, 0]])

    vertical_edges = cv2.filter2D(img, -1, vertical_mask)
    horizontal_edges = cv2.filter2D(img, -1, horizontal_mask)
    
    return vertical_edges + horizontal_edges

laplacian_img = mylaplacian(img)
sobel_img = cv2.Sobel(img, cv2.CV_64F, 1, 1, ksize=5)

# Display the results
plt.figure(figsize=(10,10))
plt.subplot(131), plt.imshow(img, cmap='gray'), plt.title('Original')
plt.subplot(132), plt.imshow(laplacian_img, cmap='gray'), plt.title('Laplacian Filter')
plt.subplot(133), plt.imshow(sobel_img, cmap='gray'), plt.title('Sobel Filter')
plt.show()
```
